# Нахождение корней уравнения |sinx| = kx

Зададим функцию, соответствующую данному уравнению, а так же функции поиска корня делением отрезка пополам для возрастающих и убывающих участков

```python
EPS = 10 ** (-8)

import math
import matplotlib
import matplotlib.pyplot as plt
import numpy

def f(x, k):
    return abs(math.sin(x)) - k * x

def left_sollution(f, a, b, k):
    while b - a > EPS:
        if f((a + b) / 2, k) <= 0:
            a = (a + b) / 2
        else:
            b = (a + b) / 2
    return a

def right_sollution(f, a, b, k):
    while b - a > EPS:
        if f((a + b) / 2, k) >= 0:
            a = (a + b) / 2
        else:
            b = (a + b) / 2
    return a
```

Найдем решения и обозначим их на графике для k = 0.1

```python
k = 0.1
x_max = 1 / k
critical_number = int(x_max / math.pi + math.pi / 4)
results = []

for i in range(critical_number):
    results.append(left_sollution(f, i * math.pi, (i + 0.5) * math.pi, k))
    results.append(right_sollution(f, (i +0.5) * math.pi, (i + 1) * math.pi, k))

figure, regular_plot = plt.subplots()
x = numpy.arange(-0.5, x_max + 0.5, 0.01)
y1 = [abs(math.sin(a)) for a in x]
y2 = [k * a for a in x]
y_results = [k * a for a in results]
regular_plot.plot(x, y1)
regular_plot.plot(x, y2)
regular_plot.plot(results, y_results, 'o')
plt.show()
print(' '.join(map(str, results)))
```
