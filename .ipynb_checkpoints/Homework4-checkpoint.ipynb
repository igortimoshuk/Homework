# Исследование расхождения значений реальной и численной производных (Тимошук Игорь)

Введем функции дифференцирования и дифференцируемую функцию

```python
import numpy
import matplotlib
import matplotlib.pyplot as plt

def different_1(f, x, h):
    result = (f(x + h) - f(x)) / h 
    return result

def different_2(f, x, h):
    result = (f(x + h) - f(x - h)) / (2 * h) 
    return result

def f(x):
    return x ** 11
```

Введем функцию создание логарифмической шкалы

```python
def x_log2(start, finish, accuracy):
    result = []
    for i in range(start, finish):
        result.append(10 ** (-i / accuracy))
    return result
```

```python
%matplotlib notebook
```

Рассмотрим разность между значением, полученным аналитически и вычисленным одноточечной схемой в точке x = 4 при разных выбранных h.

```python
fig, ax = plt.subplots()
ky = []
kx = x_log2(1, 256, 8)
for x in kx:
    ky.append(abs(11 * 4 ** 10 - different_1(f, 4, x)))
ax.loglog(kx, ky, 'o--')
plt.grid(True)
plt.show()
```

Можно заметить, что при h <= 10 ** -16 численно полученная производная обращается в 0, так как из-за точности округления берется разность значения функций в одной и той же точке. Поэтому в дальнейшем эти значения h рассматривать не будем.

Теперь рассмотрим значения производных, вычисленных по обоим схемам (двухточечная - оранжевый цвет)

```python
fig, ax = plt.subplots()
ky1 = []
ky2 = []
kx = x_log2(1, 128, 8)
for x in kx:
    ky1.append(abs(11 * 4 ** 10 - different_1(f, 4, x)))
    ky2.append(abs(11 * 4 ** 10 - different_2(f, 4, x)))
ax.loglog(kx, ky1, 'o--')
ax.loglog(kx, ky2, 'o-')
plt.grid(True)
plt.show()
```

Двухточечная схема, как и ожидалось, дает результат более чем на два порядка точнее при оптимальном h.
При дальнейшем уменьшении h погрешность, полученная из разности двух больших чисел в знаменателе производной, при делении на h возрастает прямо пропорционально первой степени h. (Это можно увидеть и по наклону левой части графика)
для одноточечной схемы правая часть графика так же линейно зависит от h, а угол наклона двухточечной схемы в 2 раза больше, что подтверждает её квадратичную зависимость от h.
